```python
import shutil
import os
import pandas as pd

path = './data/15509_18022/'
files = os.listdir(path)
new_path = './data/data/'

lists = pd.read_csv('data/result_words.csv', header = None)
lists.iloc[:,5] = lists.iloc[:,5] + '.mp4'
mp4 = lists.iloc[:,5]
```

```python
for file in files:
    for video in mp4:
        if video in file:
            shutil.move(path + file, new_path + file)
            print('{} has been moved to new folder!'.format(file))
```
